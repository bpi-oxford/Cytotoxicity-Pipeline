```python
# Paths for your images (update these with your actual paths)
#replicate_paths = ['path/to/replicate1/*.tif', 'path/to/replicate2/*.tif', 'path/to/replicate3/*.tif']
replicate_paths = [PI_IMAGE_PATH_PATTERN, PI_IMAGE_PATH_PATTERN_2]

gamma = 0.5
percentile = 1

# Function to preprocess images and compute mean intensity
def process_images(image_path_pattern):
    # Load images
    pi_image = dask_image.imread.imread(image_path_pattern)
    max_frame = pi_image.shape[0]  # Dynamically get the number of frames
    
    # Apply gamma correction
    gamma_corr = exposure.adjust_gamma(pi_image, gamma=gamma, gain=1)
    
    # Compute intensity scaling percentiles
    p2, p98 = np.percentile(gamma_corr.compute(), (percentile, 100 - percentile))
    
    # Rescale intensity
    rescaled_image = exposure.rescale_intensity(gamma_corr, in_range=(p2, p98))
    rescaled_image = exposure.rescale_intensity(rescaled_image, in_range='uint16', out_range='uint8')
    
    # Compute mean intensity per frame
    mean_intensity = da.mean(rescaled_image, axis=(1, 2)).compute()
    time = np.arange(len(mean_intensity))
    
    return pd.DataFrame({'time': time, 'mean_intensity': mean_intensity, 'max_frame': max_frame})

# Function to normalize intensities (Min-Max Scaling)
def normalize_intensity(df):
    min_val = df['mean_intensity'].min()
    max_val = df['mean_intensity'].max()
    df['mean_intensity_normalized'] = (df['mean_intensity'] - min_val) / (max_val - min_val)
    return df

# Process each replicate and normalize
replicates = []
for i, path in enumerate(replicate_paths, start=1):
    df = process_images(path)
    df['replicate'] = i
    df = normalize_intensity(df)  # Normalize the intensities
    replicates.append(df)

# Find the longest time series across replicates
max_time = max([df['time'].max() for df in replicates])

# Pad time series with NaN to ensure equal length
padded_replicates = []
for df in replicates:
    missing_frames = max_time - df['time'].max()
    if missing_frames > 0:
        padding = pd.DataFrame({
            'time': np.arange(df['time'].max() + 1, max_time + 1),
            'mean_intensity': np.nan,
            'mean_intensity_normalized': np.nan,
            'replicate': df['replicate'].iloc[0]
        })
        df = pd.concat([df, padding], ignore_index=True)
    padded_replicates.append(df)

# Combine all replicates
combined_data = pd.concat(padded_replicates)

# Compute mean and SEM for normalized intensities across replicates
summary = combined_data.groupby('time')['mean_intensity_normalized'].agg(['mean', 'sem']).reset_index()

# Plot the normalized results with individual replicates
plt.figure(figsize=(10, 6))

# Plot individual normalized means for each replicate
for i, df in enumerate(padded_replicates, start=1):
    plt.plot(df['time'], df['mean_intensity_normalized'], label=f'Replicate {i}', linestyle='--', alpha=0.7)

# # Plot the mean  SEM
# plt.errorbar(summary['time'], summary['mean'], yerr=summary['sem'], fmt='-o', capsize=5, label='Mean  SEM', color='black', linewidth=2)

# Add labels and legend
plt.title('Normalized Mean Intensity Over Time (Across Replicates)', fontsize=14)
plt.xlabel('Time (frames)', fontsize=12)
plt.ylabel('Normalized Mean Intensity', fontsize=12)
plt.legend()
plt.grid(True)
plt.show()




```

Ultrack reassigns the label ids, we need to remap back to the original segmentaiton ids by position matching

```python
from sklearn.neighbors import KDTree

cancer_df = features["cancer"]

# iterate over frames
cancer_df_remapped = []
for t in range(FRAMES):
    cancer_t_df = cancer_df[cancer_df["frame"]==t]
    tracks_t_df = tracks_df[tracks_df["t"]==t]
    points_cancer = cancer_t_df[["y","x"]].values
    points_ultrack = tracks_t_df[["x","y"]].values

    # build KD-Tree for each point in ultrack df
    kdtree_ultrack = KDTree(points_ultrack)

    # query closest point for each point in t cell df
    dist, indices = kdtree_ultrack.query(points_cancer,k=1)

    dist = [d[0] for d in dist]
    indices = [i[0] for i in indices]

    remapped_track_ids = pd.DataFrame(tracks_t_df[["track_id","id"]].iloc[indices])
    remapped_track_ids["dist_from_ultrack"] = dist

    cancer_t_df.reset_index(drop=True,inplace=True)
    remapped_track_ids.reset_index(drop=True,inplace=True)
    cancer_t_df_remapped = pd.concat([cancer_t_df,remapped_track_ids],axis=1)

    cancer_df_remapped.append(cancer_t_df_remapped)

cancer_df_remapped = pd.concat(cancer_df_remapped)

track_pair_counts = cancer_df_remapped.groupby(["frame","track_id"]).size().reset_index(name="count")
filtered_rows = track_pair_counts[track_pair_counts['count'] != 1]

sorted_df_cancer = cancer_df_remapped.sort_values(by=["frame","track_id","dist_from_ultrack"])
sorted_df_cancer = sorted_df_cancer.drop_duplicates(subset=["frame","track_id"], keep="first")

sorted_df_cancer["tracked frames"] = sorted_df_cancer['frame'].groupby(sorted_df_cancer['track_id']).transform('count')
sorted_df_cancer['tracked length'] = sorted_df_cancer.groupby('track_id')['frame'].transform("max") - sorted_df_cancer.groupby('track_id')['frame'].transform("min") +1 

# filter out short tracks
min_track_length = 50
tracks_df_filtered_cancer = sorted_df_cancer[sorted_df_cancer["tracked length"] > min_track_length]

# filter by tracked average cell size
mean_transform = lambda x: x.mean()
tracks_df_filtered_cancer['mean tracked cell size'] = tracks_df_filtered_cancer.groupby("track_id")["size"].transform(mean_transform)

area_percentiles = tracks_df_filtered_cancer.sort_values("track_id")['mean tracked cell size'].quantile(0.95)

# filter out the cells with average area larger than threshold
tracks_df_filtered_cancer = tracks_df_filtered_cancer[tracks_df_filtered_cancer["mean tracked cell size"]<area_percentiles]
tracks_df_filtered_cancer
```
